```python
from iota import Iota, Transaction, TryteString
from json import loads
from pprint import pprint
from pendulum import from_timestamp, now
from prettytable import PrettyTable
```

```python
deposit_addr = 'MPUKMWNFTYFRLJDE9ZWJY9JPKVEIIKDOWANMJIHJJWPOINFRXKVLWOUHFCMCWLO9GAAWDRWGXMTKFCIZDDQTTHNERC'
iota_node = 'https://potato.iotasalad.org:14265'
```

```python
iota = Iota(iota_node)
iota.get_node_info()
```

```python
tss = iota.find_transactions(addresses=[deposit_addr])
```

```python
states = iota.get_latest_inclusion(tss["hashes"])
```

```python
ths = iota.get_latest_inclusion(tss.get("hashes", []))
```

```python
active_transactions = list(filter(lambda ts: ths["states"][ts], ths["states"]))
# active_transactions = ths["states"]
active_transactions
```

```python
pt = PrettyTable()
pt.field_names = ["time", "value", "data", "confirmed", "hash"]
```

```python
for t in active_transactions:
    try:
        t_bytes = bytes(t)
        t_trytes = str(iota.get_trytes([t_bytes])['trytes'][0])
        transaction = Transaction.from_tryte_string(t_trytes)
        t_timestamp_clean = from_timestamp(transaction.attachment_timestamp / 1000)
        t_data = transaction.signature_message_fragment.decode()
        t_status = transaction.is_confirmed
        pprint(transaction.as_json_compatible())
    except TypeError:
        continue
    try:
        json_data = loads(t_data)
    except:
        json_data = str(t_data)
    pt.add_row([t_timestamp_clean, transaction.value, json_data, t_status, f"{transaction.hash[:12]}..."])
```

```python
pt.sortby = "time"
print(pt)
```
